```python
app_train.describe()
```

```python
def remove_outlier(dataset,col):
    mean = np.mean(dataset[col], axis=0)
    sd = np.std(dataset[col], axis=0)
    final_list=[]
    final_list = [x for x in dataset[col] if (x > mean - 2 * sd)]
    final_list = [x for x in final_list if (x < mean + 2 * sd)]
    return(final_list)
```

```python
fig = plt.figure(figsize=(15,3))

ax1 = fig.add_subplot(1,3,1)
ax1.hist(remove_outlier(app_train,'AMT_ANNUITY'))
ax1.set_title("Plot of AMT_ANNUITY")

ax2 = fig.add_subplot(1,3,2)
ax2.hist(remove_outlier(app_train,'AMT_CREDIT'))
ax2.set_title("Plot of AMT_CREDIT")

ax3 = fig.add_subplot(1,3,3)
ax3.hist(remove_outlier(app_train,'AMT_INCOME_TOTAL'))
ax3.set_title("Plot of AMT_INCOME_TOTAL")
```
